```python
%matplotlib inline
import matplotlib.pyplot as plt
plt.rcParams['figure.dpi'] = 72
plt.figure(figsize=(8, 6))

import numpy as np
from scipy.stats import t
```

```python
dqn_dir = '../output/final_dqn_bet_size_runs'
fixed_dir = '../output/final_fixed_bet_size_runs'
random_dir = '../output/final_random_bet_size_runs'
```

```python
dqn_bet_size_prefix = '2024_05_03_07_42_03'
fixed_bet_size_prefix = '2024_05_02_20_37_49'
random_bet_size_prefix = '2024_05_02_19_59_11'
```

```python
agents_balance = np.load(f'../output/{dqn_bet_size_prefix}_agents_balance.npy')

average_balance = np.mean(agents_balance, axis=0)

plt.plot(average_balance, label='Average Agent Balance')

plt.title('Average DQN Agent Balance Over Time')
plt.xlabel('Episodes')
plt.ylabel('Average Balance')
plt.legend(loc='lower left')
plt.grid(True)
plt.show()

final_balances = agents_balance[:, -1]
mean = np.mean(final_balances)
std_dev = np.std(final_balances, ddof=1)

df = len(final_balances) - 1

confidence_level = 0.95
t_crit = t.ppf((1 + confidence_level) / 2, df)
margin_error = t_crit * (std_dev / np.sqrt(len(final_balances)))

conf_interval = (mean - margin_error, mean + margin_error)

print(f'Mean: {mean}')
print(f'Standard Deviation: {std_dev}')
print(f'95% Confidence Interval: {conf_interval}')
```
